```python
#Initializing environment
env = gym.make('Humanoid-v2')
So = env.reset()
```

```python
#Sampling Action Space
A = []
a = 1000
global A_count
A_count  = np.zeros(a)

for _ in range(a):
    A.append(env.action_space.sample())
    
def Resample_action_space(A):
    dropout_rate = 0.3
    sorted_actions = sorted(list(zip(A,A_count.tolist())), key = lambda x:x[1], reverse = True)
    B = [action[0] for action in sorted_actions[0:(int(len(A)*(1-dropout_rate)))]] + [env.action_space.sample() for _ in range(int(len(A)*dropout_rate))]
    return B
```

```python
#Creating the model for q(S,a)
Q_model = Sequential()
Q_model.add(Dense(400,input_shape = np.append(So,A[0]).shape, activation = 'relu' ))
Q_model.add(Dropout(0.5))
Q_model.add(Dense(100, activation = 'relu'))
Q_model.add(Dropout(0.5))
Q_model.add(Dense(30, activation = 'relu'))
Q_model.add(Dense(1, activation= 'linear'))

#adam = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
Q_model.compile(loss="mse", optimizer= "sgd")
print(Q_model.summary())
```

```python
#The e-greedy policy

def e_greedy_Policy(s, A, q_model):
    
    e = 0.25
    best_action = 0
    best_prediction = q_model.predict(np.array([np.append(s,A[0])]))[0][0]
    
    for i in range(len(A)):
        value_prediction = q_model.predict(np.array([np.append(s,A[i])]))[0][0]
        if(value_prediction > best_prediction):
            best_action = i
            best_prediction = value_prediction
    
    if(np.random.rand() <= e):
        best_action = np.random.randint(len(A))
    
    A_count[best_action] = A_count[best_action] + 1
    return A[best_action]
```

```python
#Sarsa Parameters
lamb = 0.9
s = So
action = e_greedy_Policy(s, A, Q_model)
number_of_episodes = 0
total_reward=0
```

```python
while(True):
    
    #renders action 
    env.render()
    
    #steps
    state, reward, done, info = env.step(action)
    
    #selects new action acording to the policy
    next_action = e_greedy_Policy(state, A, Q_model)
    
    #computes the target for Q approximation
    target = reward + lamb * Q_model.predict(np.array([np.append(state,next_action)]))[0][0]
    
    #adjusts the q value
    Q_model.fit(np.array([np.append(s,action)]) ,  np.array([target]), verbose = 0)
    
    #adjusts new state
    s = state
    action = next_action 
    total_reward = total_reward + reward
    
    if(done):
        
        s = env.reset()
        action = e_greedy_Policy(s, A, Q_model)
        print(total_reward)
        total_reward = 0
        
        #Resample action space
        number_of_episodes = number_of_episodes + 1
        if(number_of_episodes%10==0):
            A = Resample_action_space(A)
            A_count = np.zeros(len(A))
```
